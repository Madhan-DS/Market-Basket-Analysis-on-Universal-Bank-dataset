```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori,association_rules
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("universalBank.CSV")
```

```python
df.head()
```

```python
df1=df.iloc[:,9:14]
```

```python
df1 
```

```python
freq_items=apriori(df1,min_support=0.01,use_colnames=True,verbose=1)
freq_items.head(7)
```

## itemsets will only have a support metric
## rules will have confidence and lift along with support

```python
freq_items
```

```python
rules=association_rules(freq_items,metric='confidence',min_threshold=0.85)
```

# for evaluation or goodness of rules the metrics looked after are         support,confidence ,(lift>1) !!
## support minimum and high are also acceptable
## confidence has to be high
## lift>1 (min. requirement)

```python
rules
```

```python
rules.lift
```

```python
rules[rules.lift<1]
## understand the difference between this and above one
```

```python
rules[rules.lift>1]
## is same as df[df.reqindex]
## these are the well associated rules with their lift>1
```
